```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("output.json", orient='index')
df = df.rename(columns={0: "Duration (ms)"}).sort_index()
df["Duration (ms)"] = df["Duration (ms)"].apply(lambda x: pd.to_timedelta(x).total_seconds() * 1e3)

df.plot.line(legend=None)
plt.xlabel("# of Competing First() calls")
plt.ylabel("Duration (ms)")
```
